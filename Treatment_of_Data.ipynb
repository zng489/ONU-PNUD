```python
import pandas as pd
import sys
from datetime import datetime

import time
start_time = time.time()


try:
    #df = pd.DataFrame()
    print("Aguarde!")

    df1 = pd.read_csv('vagasCertas2021-05-10-11-54.csv',error_bad_lines=False, sep=';')
    # df1.rename({'Unnamed: 0': 'id'}, axis=1, inplace=True)
    #del df1['Unnamed: 0']
    df2 = pd.read_csv('vagasCertas2021-05-10-12-47.csv',error_bad_lines=False, sep=';')
    # df1.rename({'Unnamed: 0': 'id'}, axis=1, inplace=True)
    #del df2['Unnamed: 0']
    df3 = pd.read_csv('vagasCertas2021-05-10-13-41.csv',error_bad_lines=False, sep=';')
    # df1.rename({'Unnamed: 0': 'id'}, axis=1, inplace=True)
    df4 = pd.read_csv('vagasCertas2021-05-10-14-35.csv',error_bad_lines=False, sep=';')
    #df1.rename({'Unnamed: 0': 'id'}, axis=1, inplace=True)

    df = pd.concat([df1,df2,df3,df4],join='inner')

    df['Codigo'] = df['URL'].str.slice(35,)
    df['Data do scraping'] = datetime.now().strftime('%d/%m/%Y')

    del df['Unnamed: 0']
    df['Vaga'] = df['Vaga'].str.slice(2, -2)
    df['Salário'] = df['Salário'].str.slice(2, -2)
    df['Localidade'] = df['Localidade'].str.slice(2, -2)
    df['Data'] = df['Data'].str.slice(2, -2)
    df['Conteúdo'] = df['Conteúdo'].str.slice(2, -2)
    df['Benefícios'] = df['Benefícios'].str.slice(2, -2)
    df['Regime'] = df['Regime'].str.slice(2, -2)
    df['Adicional'] = df['Adicional'].str.slice(2, -2)
    df['Dados da empresa'] = df['Dados da empresa'].str.slice(2, -2)

    #colunas = ['URL','Codigo','Vaga', 'Salário', 'QtdVagas', 'Localidade', 'Data', 'Conteúdo', 'Benefícios', 'Horário', 'Regime','Adicional', 'Dados da empresa']

    df.rename({'URL': 'Url','Codigo':'Código','Vaga':'Nome_Vaga','QtdVagas': 'Vagas','Dados da empresa':'Dados da Empresa'}, axis=1, inplace=True)

    for i in df.columns:
        df[i] = df[i].astype(str).str.replace(r"\\r", '', regex=True)
        df[i] = df[i].astype(str).str.replace(r"\\n", '', regex=True)
        df[i] = df[i].astype(str).str.replace(r"', '', '", '', regex=True)
        df[i] = df[i].astype(str).str.replace(r"', '", '', regex=True)
        df[i] = df[i].astype(str).str.replace('=', '')

    longDf = df.stack()
    longDf = pd.DataFrame(longDf)
    longDf = longDf.reset_index()

    longDf.rename(columns={'level_0': 'ID','level_1':'TAG',0:'TEXTO'}, inplace=True)

    nome = 'DF_Long2.csv'
    longDf.to_csv(nome, sep=';', encoding="utf-8-sig", index=False)

except:
    print('Erro no Programa!', sys.exc_info()[0])
    raise
finally:
    # fim = time.time()
    print("--- %s seconds ---" % (time.time() - start_time))
    print("Finalizado")
```
