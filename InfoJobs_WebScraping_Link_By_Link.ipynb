```python
import csv

from bs4 import BeautifulSoup
import requests

dados_infojobs = []

for i in range(1, 3):
    urls = "https://www.infojobs.com.br/empregos.aspx?Page={numbers_pages}".format(numbers_pages=i)
 
    option = webdriver.ChromeOptions()
    option.add_argument('headless')
    chromedriver = 'C:/Users/Zng/Desktop/Scrapy/chromedriver_win32/chromedriver'

    #driver = webdriver.Chrome(chromedriver)
    driver = webdriver.Chrome(chromedriver,options=option)

    sleep(2)

    driver.get(urls)
    
    sleep(2)
    
    try:
        botao_FECHAR_1 = driver.find_element_by_xpath("/html/body/form/div[10]/div/div[3]/a").click()
    except:
        pass
#####################################################################

    elements = driver.find_elements_by_xpath('//div[@class="element-vaga unstyled noborder    lider   js_detailGrid" or @class="element-vaga unstyled   js_VacancyLimited    js_detailGrid premiumLimited" or @class="element-vaga unstyled      hiddenCompany js_detailGrid" or @class="element-vaga unstyled       js_detailGrid" or @class="element-vaga unstyled noborder    lider  hiddenCompany js_detailGrid"]')    

    for element in elements:
    
        codigo = element.get_attribute('id')
    
        links = element.find_element_by_xpath('./div/div[2]/a[@class="vagaTitle js_vacancyTitle"]').get_attribute('href')

        data = element.find_element_by_xpath('/html/body/form/section/div/div[1]/section[2]/div[2]/div[2]/div/div[4]/div[2]/p[1]/span[1]').text
    
#river.find_element_by_xpath('/html/body/form/section/div/div[1]/section[2]/div[2]/div[3]/div/div[4]/div[2]/p[1]/span[1]').text
#####################################################################    
#Scrape de cada pagina clicado nos links 

        
        bs4_urls = links 
        r = requests.get(bs4_urls).text  # content (would be preferred for "binary" filetypes, such as an image or PDF file)
        soup_h = BeautifulSoup(r, 'xml')
        
        try:
            nome_vaga = soup_h.find('div',{'class':"header-content-left"}).find('h1').getText().replace('\n','').replace('   ','')
        except:
            nome_vaga = '-'
            
        try: 
            dados_da_empresa = soup_h.find('h1',{'class':"header-name"}).getText()
        except:
            dados_da_empresa = '-'
            
        try:
            vagas = soup_h.find('span',{'class':"advisor-number-candidate js_numberCandidate"}).getText().replace('\n','').replace('   ','')
        except:
            vagas = '-'
            
        try:
            descricao = soup_h.findAll('ol',{'class':"descriptionItems"})[0].getText().replace('\n','').replace('   ','')        
        except:
            descricao = '-'
            
        try:
            exigencias = soup_h.findAll('ol',{'class':"descriptionItems"})[1].getText().replace('\n','').replace('   ','')
        except:
            exigencias = '-'
            
        try:
            beneficios = soup_h.findAll('ol',{'class':"descriptionItems"})[2].getText().replace('\n','').replace('   ','')
        except:
            beneficios = '-'
            
        try:
            localidade = soup_h.find('span',{'id':"ctl00_phMasterPage_cVacancySummary_litLocation"}).getText()
        except:
            localidade = '-'
            
        try:
            salario = soup_h.find('span',{'id':"ctl00_phMasterPage_cVacancySummary_litSalary"}).getText().replace('R$','')
        except:
            salario = '-'
            
        try:
            regime = soup_h.find('span',{'id':"ctl00_phMasterPage_cVacancySummary_litContractType"}).getText()
        except:
            regime = '-'
            
        try:
            jornada = soup_h.find('span',{'id':"ctl00_phMasterPage_cVacancySummary_litWorkingHours"}).getText()
        except:
            jornada = '-'
        
        data_scraping = datetime.datetime.now().strftime('%d/%m/%Y')

        
        dados = {
                'Url':links,
                'Código':codigo,
                'Nome_Vaga':nome_vaga,
                "Salário":salario ,
                "Vagas":vagas,
                "Localidade":localidade,
                "Data":data,
                #Data de expiração":data_expiracao,
                "Conteúdo":descricao,
                #"Benefícios":
                #'Horário':
                'Regime':regime,
                'Jornada':jornada,
                #'Adicional':
                'Dados da Empresa':dados_da_empresa,
                #Pcd':pcd,
                "Data do scraping":data_scraping}
             
        
        dados_infojobs.append(dados)  
        
        
        #rint('Scraping....')
        
#   TrabalhaBrasil = pd.DataFrame(Dados_TrabalhaBrasil) 
#   TrabalhaBrasil = TrabalhaBrasil.fill()
#   df = TrabalhaBrasil.stack()
#   df =  df.reset_index()
#   df = df.rename(columns={'level_0': 'ID','level_1':'TAG',0:'TEXTO'}, inplace=True)
#   df.to_csv('file.csv', index=False)
      


df = pd.DataFrame(dados_infojobs)
df.to_csv('D.csv')
```
